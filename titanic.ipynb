```python
import pandas as pd
```

```python
titanic_import_df = pd.read_csv('titanic_train.csv')
```

```python
titanic_import_df #불러온 파일 보기  Pclass : 선실 등급  sibsp : 형제자매  Parch : 부모, 자식  Fare : 티켓 요금  cabin : 사물함  Embarked : 출항지
```

```python
titanic_df= titanic_import_df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']] #필요한 X
titanic_label=titanic_import_df['Survived']  # 필요한 y
```

```python
titanic_df
```

```python
titanic_df.isna().sum() #결측값 조사
```

```python
titanic_df['Age']=titanic_df['Age'].fillna(titanic_df['Age'].mean()) #나이는 중간 값으로 대체
```

```python
titanic_groupbyEmbarked = titanic_df.groupby('Embarked').count()
titanic_groupbyEmbarked
```

```python
titanic_df['Embarked'].fillna('S',inplace=True) # 출발지는 제일 많은 S로 대체
```

```python
titanic_df.isna().sum()
```

```python
titanic_df['Sex2Num']=titanic_df['Sex'].apply(lambda x : 1 if x == 'male' else 0) #문자열 -> 숫자로 변환 feature encoding하면됨
```

```python
titanic_df.drop(columns='Sex',inplace=True)
```

```python
titanic_df
```

```python
titanic_df['Embarked2Num']= titanic_df['Embarked'].apply(lambda x: 0 if x =='Q' else (1 if x=='C' else 2))
```

```python
titanic_df.drop(columns='Embarked',inplace=True)
```

```python
titanic_df
```

```python
titanic_data=titanic_df.values 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(titanic_data,titanic_label,test_size=0.2,random_state=1) #데이터 셋 나눔
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=1) #학습할 모델
```

```python
model.fit(X_train,y_train) #학습
```

```python
pred = model.predict(X_test) #예측
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
